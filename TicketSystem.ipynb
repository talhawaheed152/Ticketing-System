```python
import csv

class TicketSystem:
    Name=" "
    
    def __init__(self):
        x=" "
        while(x!='exit'):
            x=input(" Welcome to The Ticketing Sysyem:\n\n Select Login / Sign Up / Exit:\n ")
            x=x.lower()
            if x=='login':
                self.Login()
            elif x=='signup':
                self.SignUp()
    
    def SignUp(self):
        x=input(" Enter Name: ")
        y=input(" Enter Password: ")
        z=input(" Enter Email: ")
        str= x+' , '+y+' , '+z+'\n'

        with open('credentials.sql','a') as file:
            file.write(str)
            file.close()
        print(" Signed Up Successfully !")
        print("\n Returning Back to Menu ")

    
    def Login(self):
        x=input(" Enter Name: ")
        y=input(" Enter Password: ")
        str= x+' , '+y+'\n'

        with open('credentials.sql','r') as file:
            Credential=file.readlines();
            file.close()
        for l in Credential:
            if l==str:
                print(" Logged In Successfully !")
                self.Name=x
                self.Options()
                return
        print(" Logging in Failed! Name or Password incorrect!\n ")
        
    def Options(self):
        x=input(" Please Select an Option From the Following: \n Book Tickets / Show Booked Details / Exit \n ")
        x=x.lower()
        if x=='exit':
            return 
        elif x=='show booked details':
            self.Details()
            return
        elif x=='book tickets':
            with open('movies.sql','r') as file:
                movie=file.readlines();
                file.close()
                while(1):
                    count=1
                    print(" following are the movies available: ")
                    for ik in movie:
                        j=ik.split(',')
                        print(" ",count,". ",j[0].strip())
                        print("  Movie Date: ",j[1].strip(),'\n')
                        count+=1

                    x=input(" Enter movie name:\n ")
                    y=input(" Enter movie date with format DD/MM/YYYY:\n ")
                    for i in movie:
                        j=i.split(',')
                        if x==j[0].strip():
                            print(" Tickets Available! Taking to The Booking: \n")
                            self.Booking(x,y,int(j[2].strip()))
                            return 
                    print("This movie is not available, please choose from the list !")
    
    def Booking(self,m,d,s):
        
        with open('screen.txt','r') as file:
            screen=file.readlines()
            file.close()
        for s in screen:
            print(s,'\n')
        print("--- BOOKING ---\n")
        with open('bookings.sql','r') as file:
            book=file.readlines()
            file.close()
        x=input(" Enter No. of Seats to book: ")
        y=" "
        l=[]
        si=0
        for ii in book:
            p=ii.split(',')
            size=len(p)-3
            if p[0].strip()==m and p[2].strip()==d:
                siz=0
                while siz<size:
                    p[3+siz].strip()
                    l.append(p[3+siz].strip())
                    siz+=1
        if l>=30 or len(l)+int(x)>=30:
            print("Seats Not Available! \n")
            return
        flag=0
        i=0
        seats=['A1','A2','A3','B1','B2','B3','C1','C2','C3','D1','D2','D3','E1','E2','E3','F1','F2','F3','G1','G2','G3','H1','H2','H3','I1','I2','I3','J1','J2','J3']
        while i<int(x):
            z=input(" Enter Seat Name: \n ")
            for o in l:
                if z==o:
                    flag=1
            if flag==0:
                if z in seats:
                    y=y+" , "+z
                    i+=1
                else: 
                    print(" No Such Seat Name Exists! \n")
            else:
                print(" Seat Already Taken !\n")
            flag=0
             
        str= m+' , '+self.Name+' , '+d+y+'\n'
        with open('bookings.sql','a') as file:
            file.write(str)
            file.close()
    
    def Details(self):
        x=input(" Enter Email: ")
        with open('bookings.sql','r') as file:
            book=file.readlines()
            file.close()
        x=[]
        for i in book:
            x=i.split(',')
            if x[1].strip()==self.Name:  
                print(' Movie: ',x[0].strip())
                print('\n Booked by: ',x[1].strip())
                print('\n Date: ',x[2].strip())
                size=3
                print('\n Seat(s): ')
                while size<len(x):
                    print(" ",x[size].strip()," ",end="")
                    size+=1
                print('\n')
        
    
x=TicketSystem()
```
